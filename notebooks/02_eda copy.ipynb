```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Plotting

```python
#Importing updated dataset from previous notebook
diabetes = pd.read_csv('/content/My-Clean-Output.csv', index_col = 0)
diabetes.head()
```

Correlation Matrix

```python
sns.heatmap(diabetes.corr(), annot = True)
plt.show();
```

Findings from correlation matrix: The strongest relationship tied to the outcome variable is glucose meaning the higher the glucose, more likely someone can get diabetes.

Although the weakest correlations tied to the outcome variable are blood pressure and Diabetes Pedigree Function (DPF), DPF have the weakest relationship with other variables.

For example, DPF and Blood Pressure have a correlation of -0.0024 meaning the likelihood of developing diabetes based on family history obviously has almost to no affect on blood pressure.

Histograms

```python
#Defining variables
X = diabetes[diabetes.columns[:-1]]
y = diabetes['Outcome']
```

```python
# Now plotting histograms
for col in X.columns:
  plt.hist(X[col])
  plt.xlabel(col)
  plt.show();
```

Glucose and Blood Pressure seem to be normally distrubted while the rest is not.

Boxplots

```python
for col in X.columns:
  sns.boxplot(X[col])
  plt.show();
```

According to the box plots, glucose is the only variable that has no outliers.

The boxplot for pregnancies have only 3 outliers, but the histogram still suggests that it is not normally distributed.

The boxplot for blood pressure has a few outliers, but the histogram for it is normally distributed and is bilogically possible to have those values.
